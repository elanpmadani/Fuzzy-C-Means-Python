```python
def matrikskeanggotaan():
    matriksu = list()
    for i in range(len(data)):
        randomnumber = [random.random() for x in range(c)]
        jumlah = sum(randomnumber)
        temp = list()
        for x in randomnumber:
            temp.append(round(x/jumlah,3))
        matriksu.append(temp)
    return matriksu
```

```python
def centroid(keanggotaan):
    xraised = []
    jumlahcentroid = []
    centroid = []
    for x in range(len(data)):
        raised = []
        for y in range(c):
            cie = math.pow(keanggotaan[x][y],w)
            raised.append(cie)
        xraised.append(raised) 
    for x in range(c):
        sumz = 0
        for y in range(len(data)):
            sumz += xraised[y][x]
        jumlahcentroid.append(sumz)
    for i in range(c):
        temp = []
        for x in range(len(data[0])):
            sum_u = 0
            for y in range(len(data)):
                sum_u += (xraised[y][i]*data[y][x])/(jumlahcentroid[i])
            temp.append(round(sum_u,4))
        centroid.append(temp)
    return centroid
```

```python
def obyektif(centroid,keanggotaan):
    obj = 0
    dist = []
    xraised = []
    for x in range(len(data)):
        raised = []
        for y in range(c):
            cie = math.pow(keanggotaan[x][y],w)
            raised.append(cie)
        xraised.append(raised)
    for x in range(len(data)):
        temp = []
        for y in range(c):
            sum_z = 0
            for i in range(len(data[0])):
                sum_z += math.pow(data[x][i]-centroid[y][i],2)
            temp.append(sum_z)
        dist.append(temp)
    for x in range(len(data)):
        for y in range(c):
            obj += dist[x][y]*xraised[x][y]
    return obj
```

```python
def updateMatriks(keanggotaan,centroid):
    dist = []
    for x in range(len(data)):
        temp = []
        for y in range(c):
            sum_z = 0
            for i in range(len(data[0])):
                sum_z += math.pow(data[x][i]-centroid[y][i],2)
            temp.append(round(math.sqrt(sum_z),4))
        dist.append(temp)
    distraised = []
    for i in range(len(dist)):
        temp = []
        for j in range(len(dist[0])):
            x=1/(dist[i][j]**2)
            temp.append(x)
        distraised.append(temp)
    jumlah = []
    pop = 0
    for i in range(len(distraised)):
        for j in range(len(distraised[0])):
            pop += distraised[i][j]
        jumlah.append(pop)
        pop = 0
    updatematriksu = []
    for i in range(len(distraised)):
        temp = []
        for j in range(len(distraised[0])):
            x = distraised[i][j]/jumlah[i]
            temp.append(x)
        updatematriksu.append(temp)
    return updatematriksu
```

```python
def findcluster(u):
    cluster = []
    for x in u:
        cluster.append(np.argmax(x)+1)
    return cluster
```
